```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import scipy as sp
from google.colab import files
uploaded = files.upload()
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
  orig = pd.read_csv(fn,sep='|')
df1 = orig.copy()
df1.describe()
```

```
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering

df=df1.select_dtypes(include=np.number)
df2=df.drop(columns='id').dropna(axis='columns')
df2.tail()

clustering=AgglomerativeClustering(n_clusters=5)
clust=clustering.fit(df2)

sc_clust = pd.DataFrame(clust.labels_, columns=['cluster'])
df_clust = pd.concat([df1, sc_clust], axis=1)
```
